2-D **Plots**

```python
df1=pd.read_csv("forestfire.csv")
df1
```

```python
subset3=df1.loc[210:215]
subset3
sub3=subset3[['X','Y','month','FFMC','DMC','area']]
sub3
```

```python
X = sub3['X']
Y = sub3['Y']
FFMC = sub3['FFMC']

# Create a grid of points
X_unique = np.sort(X.unique())
Y_unique = np.sort(Y.unique())
X_grid, Y_grid = np.meshgrid(X_unique, Y_unique)

# Interpolate 'FFMC' values onto the grid
FFMC_grid = np.zeros_like(X_grid, dtype=float)
for x, y, ffmc in zip(X, Y, FFMC):
    i = np.argmin(np.abs(X_unique - x))
    j = np.argmin(np.abs(Y_unique - y))
    FFMC_grid[j, i] = ffmc

# Plot contour
plt.contour(X_grid, Y_grid, FFMC_grid)
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.title('Contour Plot of FFMC')
plt.colorbar(label='FFMC')
plt.show()
```

```python
X = sub3['X']
Y = sub3['Y']
FFMC = sub3['FFMC']

# Create a grid of points
X_unique = np.sort(X.unique())
Y_unique = np.sort(Y.unique())
X_grid, Y_grid = np.meshgrid(X_unique, Y_unique)

# Interpolate 'FFMC' values onto the grid
FFMC_grid = np.zeros_like(X_grid, dtype=float)
for x, y, ffmc in zip(X, Y, FFMC):
    i = np.argmin(np.abs(X_unique - x))
    j = np.argmin(np.abs(Y_unique - y))
    FFMC_grid[j, i] = ffmc

# Plot pcolormesh
plt.pcolormesh(X_grid, Y_grid, FFMC_grid, cmap='viridis', shading='auto')
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.title('pcolormesh Plot of FFMC')
plt.colorbar(label='FFMC')
plt.show()
```
