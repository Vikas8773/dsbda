```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('forestfire.csv')
df
```

#Scatter Plot

```python
plt.scatter(df['temp'],df['RH'])
plt.xlabel('Temperature')
plt.ylabel("Relative Humidity")
plt.title("Temperature v/s Relative Humidity")
plt.show()
```

#Line Plot

```python
# df = pd.read_csv("forestfire.csv")

# Convert the 'month' column to categorical to preserve the order
df['month'] = pd.Categorical(df['month'],categories=['jan', 'feb', 'mar', 'apr', 'may', 'jun','jul', 'aug', 'sep', 'oct', 'nov', 'dec'],ordered=True)

# Group by month and calculate the average values of temperature, relative humidity, and rainfall
monthly_avg_df = df.groupby('month')[['temp', 'RH', 'rain']].mean()

# Plot the seasonal patterns
plt.figure(figsize=(10, 6))
monthly_avg_df.plot(marker='o', markersize=8)
plt.title('Seasonal Patterns in Environmental Conditions')
plt.xlabel('Month')
plt.ylabel('Average Value')
plt.grid(True)
plt.legend(['Temperature', 'Relative Humidity', 'Rainfall'])
plt.xticks(range(12), monthly_avg_df.index)
plt.tight_layout()
plt.show()
```

#Bar

```python
df['day']=pd.Categorical(df['day'],categories=['mon','tue','wed','thu','fri','sat','sun'],ordered=True)
day_avg_df=df.groupby('day')[['temp']].mean()
plt.figure(figsize=(8, 6))
# Plotting bar chart
plt.bar(day_avg_df.index, day_avg_df['temp'], color='skyblue')
plt.title('Average Temperature for Each Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Temperature')
plt.grid(True)
# plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
